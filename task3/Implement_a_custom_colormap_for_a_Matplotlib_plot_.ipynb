```python
import numpy as np


x = np.linspace(-3, 3, 300)
y = np.linspace(-3, 3, 300)
X, Y = np.meshgrid(x, y)


Z = np.sin(X**2 + Y**2)
```

```python
import matplotlib.pyplot as plt
from matplotlib.colors import LinearSegmentedColormap


colors = ["#00008B", "#00CED1", "#FFD700", "#FF4500"]
custom_cmap = LinearSegmentedColormap.from_list("my_custom_cmap", colors, N=256)
```

```python
plt.figure(figsize=(8, 6))
plt.imshow(Z, extent=[-3, 3, -3, 3], origin='lower', cmap=custom_cmap)
plt.title("2D Sine Function Heatmap with Custom Colormap")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.colorbar(label="sin(X² + Y²)")
plt.tight_layout()
plt.show()
b
```
