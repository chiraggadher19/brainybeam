```python
import pandas as pd


data = {
    'User_Gender': ['Male', 'Female', 'Male', 'Female', 'Female', 'Male', 'Female', 'Male', 'Male', 'Female'],
    'Course_Category': ['Programming', 'Design', 'Marketing', 'Programming', 'Design', 'Marketing', 'Programming', 'Marketing', 'Design', 'Programming'],
    'Device_Used': ['Laptop', 'Mobile', 'Tablet', 'Laptop', 'Tablet', 'Mobile', 'Laptop', 'Mobile', 'Tablet', 'Laptop'],
    'Payment_Status': ['Success', 'Failed', 'Success', 'Success', 'Failed', 'Success', 'Success', 'Failed', 'Success', 'Success']
}


df = pd.DataFrame(data)
df
```

```python

cross_tab = pd.crosstab(index=df['User_Gender'],
                        columns=[df['Course_Category'], df['Payment_Status']],
                        margins=True)

cross_tab
```

```python

cross_tab.to_csv("task1_crosstab_output.csv")
```
